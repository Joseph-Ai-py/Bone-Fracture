# 여러 골절 데이터
* 논문 참고 : https://www.nature.com/articles/s41597-023-02432-4
* Github : https://github.com/XLR8-07/FracAtlas
* 가장 흔하게 사용되는 골절 데이터는 MURA, FracAtlas




* **MURA**
    * 40,561개의 2D 근골격계 X-ray 이미지
    * 팔꿈치, 손가락, 팔뚝, 손, 상완골, 어깨, 손목
    * 정상, 비정상으로만 분류 -> <U>Classification 용이</U>
    
    
* <span style="color: #2D3748; background-color:#fff5b1;">**FracAtlas**</span>
    * 4,083개의 CT 이미지
    * 손, 어깨, 다리, 엉덩이
    * 정상, 비정상 분류 / <span style="color: #2D3748; background-color:#fff5b1;"> 골절 부분 라벨링 되어있다.</span> -> <U>Classification, <span style="color: #2D3748; background-color:#fff5b1;">Segmentation 용이</span></U>
    
    
* 어린 대상자(0~7세)의 경우 관절 근처 뼈 끝은 뼈 밀도가 부족하여 분리된 분리된 디스크 패드처럼 보일 수 있습니다. 그로인한 잘 못 된 골절 인식 가능
* 노령 환자(50세 이상)의 경우 뼈 표면이 거칠어 보일 수 있습니다 . 이로 인해 모델이 해당 뼈를 골절로 잘못 판단할 수도 있습니다. 
* 그래서 다양한 연령대의 골절 데이터를 수집할 필요가 있다.

# 파일 구조
* images 폴더 안에 Fractured, Non_fractured가 골절 이미지 데이터
* Annotations 폴더 안에 VGG_fracture_masks.json 파일이 라벨링 데이터
